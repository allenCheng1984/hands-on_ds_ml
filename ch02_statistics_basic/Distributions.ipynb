# Examples of Data Distributions

## Uniform Distribution(均勻分布)
意味著在給定的範圍內，數值出現的機率是固定的<br>
呈現的函數會是平坦的直線，因為每個值出現的機率都是一樣的

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# 取呈均勻分配的隨機數值
# ref: https://numpy.org/doc/stable/reference/random/generated/numpy.random.RandomState.uniform.html
# sample: np.random.uniform(low=0.0, high=1.0, size=None)

# 最低值 -10, 最高值 10, 總筆數 100000
values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

## Normal / Gaussian
常態分佈 / 高斯分佈

Visualize the probability density function:

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# ref: https://numpy.org/doc/stable/reference/generated/numpy.arange.html?highlight=arange#numpy.arange

# 建立一個 array, 在 -3 與 3 的之間以 0.001 的長度為間隔
x = np.arange(-3, 3, 0.001)
print(f'x lenth: {len(x)}')
print(f'x: {x}')
print(f'y: {norm.pdf(x)}')

# 使用 scipy.stats.norm.pdf 函數(即為機率密度函數, Probability Density Function)
# 來呈現常態分佈曲線

# x 軸是一連串的數值, y 軸是對應 x 值的常態分佈機率密度的函數值
plt.plot(x, norm.pdf(x))
```

Generate some random numbers with a normal distribution. "mu" is the desired mean, "sigma" is the standard deviation:

使用常態分佈來產生亂數

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0    # 代表平均數
sigma = 2.0 # 代表標準差
values = np.random.normal(mu, sigma, 10000) 
plt.hist(values, 50)
plt.show()

# 這張圖表會呈現之前看到的常態分佈，但因為是隨機數值，所以不是一條完美的曲線
# 由於機率的關係，有些值難免有波動
```

## Exponential PDF / "Power Law"
指數機率分佈與幂次定律

另一種常見的分佈函數是`指數機率分佈函數(Exponential Probability Distribution Function)`<br>
其函數值會急遽下降

這裡指的急遽下降，是形容一條曲線上在零附近的值發生的機率很高<br>
但當遠離零時，機率會大幅降低，日常生活中有許多事物符合這種情況。



```python
from scipy.stats import expon
import matplotlib.pyplot as plt

# 建立一個 array, 在 0 與 10 的之間以 0.001 的長度為間隔
x = np.arange(0, 10, 0.001)

# 使用 scipy.stats.expon.pdf 函數(即為指數機率分佈函數, Exponential Probability Distribution Function)
# 來呈現幂次定律曲線

# x 軸是一連串的數值, y 軸是對應 x 值的幂次定律機率分佈的函數值
plt.plot(x, expon.pdf(x))

```

## Binomial Probability Mass Function(二項式機率質量函數)

我們也可以對機率質量函數進行視覺化，比如二項式機率質量函數

```python
# ref: https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.binom.html
# ref: https://zh.wikipedia.org/zh-tw/%E4%BA%8C%E9%A0%85%E5%BC%8F%E5%88%86%E5%B8%83
from scipy.stats import binom
import matplotlib.pyplot as plt

# 建立一個 array, 在 0 與 10 的之間以 0.001 的長度為間隔
x = np.arange(0, 10, 0.001)

n, p = 10, 0.5 # binom.pmf 的 shape parameters 參數
# X 軸為變數 x 的資料 (0~10)
# y 軸的值是對應 x 值的 Binomial 機率質量函數的值
plt.plot(x, binom.pmf(x, n, p))
```

## Poisson Probability Mass Function(卜瓦松機率質量函數)

這個是比較特殊的分佈，它和常態分佈長的非常像，但有些許不同<br>
如果你知道給定時間內某事件發生的平均次數，<br>
就可以用`卜瓦松機率質量函數`來預測未來某個時間該事件發生特定次數的機率


Example: My website gets on average 500 visits per day. What's the odds of getting 550?

```python
# ref: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html
# ref: https://zh.wikipedia.org/wiki/%E5%8D%9C%E7%93%A6%E6%9D%BE%E5%88%86%E5%B8%83
from scipy.stats import poisson
import matplotlib.pyplot as plt

# 建立一個 array, 在 400 與 600 的之間以 0.5 的長度為間隔
x = np.arange(400, 600, 0.5)

mu = 500
plt.plot(x, poisson.pmf(x, mu))

# 舉個例子，假設有一個網站，每天平均訪客次數是 500 人次，那麼就可以用這個函數來估算某天出現另一個訪客數的機率
# 重要前提 => 假設這個資料符合常態分佈

# 由此可以推算某天有 550 個訪客的機率是 0.002 (0.2%)
```
